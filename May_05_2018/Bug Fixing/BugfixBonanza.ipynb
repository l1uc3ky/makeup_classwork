```python
#Correction: need to import dependencies 
import pandas as pd 
```

```python
# Create a reference to the CSV and import it into a Pandas DataFrame
csv_path = "EclipseBugs.csv"
# need to define dataframe 
eclipse_df = pd.read_csv(csv_path)
eclipse_df.head()
```

```python
# get reference to column names to see if number of comments is correct column name 
eclipse_df.columns
```

```python
# correction to error due to new lines keeping us from finding average 
# remove newlines from column headers 
eclipse_df = eclipse_df.rename(columns={'Bug\nID': "Bug ID",
                                        'Assignee\nReal\nName': "Assignee Real Name", 
                                        'Number of\nComments': "Number of Comments",
                                        'Reporter\nReal\nName': "Reporter Real Name",
                                        'Target\nMilestone': "Target Milestone"})
```

```python
# Finding the average number of comments per bug
average_comments = eclipse_df["Number of Comments"].mean()
average_comments
```

```python
# Grouping the DataFrame by "Assignee"
assignee_group = eclipse_df.groupby("Assignee")

# print(assignee_group["Component"].count())
print(assignee_group["Component"].value_counts())

# Count how many of each component Assignees worked on and create DataFrame
assignee_work = pd.DataFrame(assignee_group["Component"].value_counts()) #shows unique 
assignee_work.head()
```

```python
# Rename the "Component" column to "Component Bug Count"
assignee_work = assignee_work.rename(
    columns={"Component": "Component Bug Count"})
print(type(assignee_work))
assignee_work.head()
```

```python
# Find the percentage of bugs overall fixed by each Assignee
# change the data frame in total bugs to eclipse_df from assignee_work 

total_bugs = eclipse_df["Assignee"].count() 
bugs_per_user = assignee_group["Assignee"].value_counts()
print(bugs_per_user)
user_bug_percent = pd.DataFrame((bugs_per_user/total_bugs)*100) #dividing a series by a scaler 
user_bug_percent.head()
```

```python
#reset the index of assignee group so that assignee and component are columns 
assignee_work = assignee_work.reset_index()

# Merge the "Percent of Total Bugs Assigned" into the DataFrame
assignee_work = assignee_work.merge(user_bug_percent, on="Assignee")

# Remove the extra columns
assignee_work = assignee_work[["Assignee", "Percent of Total Bugs", "Component", "Component Bug Count"]]
assignee_work.head()
```

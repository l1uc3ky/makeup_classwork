```python
# Import Dependencies
import pandas as pd
```

```python
raw_data = {'Regiment': ['Nighthawks', 'Nighthawks', 'Nighthawks', 'Nighthawks', 'Dragoons', 'Dragoons', 'Dragoons', 'Dragoons', 'Scouts', 'Scouts', 'Scouts', 'Scouts'],
            'Company': ['1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd', '1st', '1st', '2nd', '2nd'],
            'Name': ['Miller', 'Jacobson', 'Ali', 'Milner', 'Cooze', 'Jacon', 'Ryaner', 'Sone', 'Sloan', 'Piger', 'Riani', 'Ali'],
            'Test Score': [25, 94, 57, 62, 70, 30, 94, 57, 62, 70, 62, 70]}
regiment_df = pd.DataFrame(raw_data)
regiment_df
```

```python
# Create the bins in which Data will be held
# Bins are 0 to 25, 25 to 50, 50 to 75, 75 to 100
bins = [0, 25, 50, 75, 100]

# Create the names for the four bins
group_names = ['Low', 'Okay', 'Good', 'Great']

#labels list has one less element than bins
```

```python
# Cut postTestScore and place the scores into bins
# tell it we want to use bins and labels should be group names 
bin_series = pd.cut(regiment_df["Test Score"], bins, labels=group_names)

#use print(type) to determine what it is 
print(type(bin_series))
print(bin_series)
```

```python
#adding a column to dataframe 
#single brackets and name the column 
# to assign something to a column use a series because print above told us it's a series 
# series can be assigned to a column 
regiment_df["Test Score Summary"] = pd.cut(
    regiment_df["Test Score"], bins, labels=group_names)
regiment_df
```

```python
# Creating a group based off of the bins
# provides you the max out of each category which maps into the numeric variable because we can take maxes of numeric variable even after it's been converted to category
regiment_groups = regiment_df.groupby("Test Score Summary")
regiment_groups.max()
```

```python
# Counting the number of results in each group
ratings_count = regiment_groups['Name'].count()
ratings_count
```
